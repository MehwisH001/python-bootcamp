## VARIABLES

- A variable stores data (like numbers, text, etc.).
- You don’t need to declare a variable type—Python understands it automatically.
- You give a variable a name and assign a value using =.

```python
# assignment creates the variable
age = 25          
name = "Mehwish"
is_learning = True
```

## Important rules (Syntax)

- Must start with letter (a–z, A–Z) or underscore _
- Can contain letters, digits (0–9), underscore
- Case-sensitive: Age ≠ age
- Cannot be a Python keyword (if, for, def, True, etc.)
- No spaces or special characters (except _)

```python
# 1. Creating variables (different types)
name = "Mehwish"          # string (text)
age = 25                  # integer (whole number)
height = 5.6              # float (decimal number)
is_learning = True        # boolean (True/False)
nothing = None            # None (no value)

# 2. Printing them
print(name)
print(age)
print(height)
print(is_learning)
print(nothing)

# 3. Type of each variable
print(type(name))         # <class 'str'>
print(type(age))          # <class 'int'>
print(type(height))       # <class 'float'>
print(type(is_learning))  # <class 'bool'>
print(type(nothing))      # <class 'NoneType'>

# 4. Changing value + dynamic typing
age = "twenty five"       
print(age)                
print(type(age))          

# 5. Multiple variables at once
x, y, z = 10, 20, "hello"
print(x, y, z)

# 6. Same value to many variables
a = b = c = 0
print(a, b, c)

```

## Good names:

## Bad names:

In conclusion, variables are a fundamental part of Python programming as they help store and manage data efficiently. Using proper naming conventions and understanding how variables work makes code more readable, organized, and easier to maintain.
