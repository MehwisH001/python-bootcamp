## Syntax

Refers to the formal rules and structure that dictate how code must be written to be valid and parsable by the interpreter or compiler.
- It's like grammar in a natural language: It defines the correct arrangement of keywords, symbols, operators, and punctuation.
- Examples in Python: Indentation for code blocks, colons after if statements, or parentheses in function calls. Invalid syntax (e.g., missing a colon) causes a SyntaxError.
- Key point: Syntax ensures the code is structurally correct but doesn't guarantee it makes sense or runs as intended.

## Semantics

Refers to the meaning and behavior of syntactically correct codeâ€”what it actually does when executed.
- It's like the meaning behind words in a sentence: Even if grammar is perfect, the intended action or logic might be wrong.
- Examples in Python: Variable assignment semantics (e.g., x = 5 creates a reference), or how operators work (e.g., + adds numbers but concatenates strings). Semantic errors might cause logical bugs, like infinite loops, without raising exceptions.
- Key point: Semantics deals with runtime behavior, logic, and interpretation, often leading to bugs that are harder to spot than syntax errors.

## Indentation in Python

- Purpose: Defines code blocks (e.g., in if, loops, functions) instead of braces {}.
- Rule: Use consistent spaces (usually 4) or tabs per level. Mix = error.
- Syntax Error: Mismatched indentation causes IndentationError.
- Semantics: Code in the same block runs together; dedent ends the block.
- Tip: Editors like VS Code auto-indent; always align visually.
