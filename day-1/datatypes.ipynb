## DATATYPES


### 1. DEFINITION
- Data type tells Python what kind of value a variable is holding and what operations you can do with it.
- Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data.
- They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.

#### 2. Importance 
- Data types ensure that data is stored in an efficient way.
- They help in performing correct operations on data.
- Proper use of data types can prevent errors and bugs in the program.

### most basic and commonly used data types 

- int → whole numbers with no decimal part (example: 5, -42, 0, 100000). Used for counting, ages, quantities, years, IDs.
- float → numbers with decimal points (example: 3.14, -0.75, 5.0). Used for measurements, prices, percentages, temperatures, scientific calculations.
- str (string) → text or sequences of characters (example: "Mehwish", 'Delhi', "hello world", ""). Used for names, messages, addresses, file paths, user input.
- bool (boolean) → only two possible values: True or False. Used for yes/no decisions, flags, conditions in if statements, loop control.
- None → represents the absence of a value (like null in other languages). Used when something has no value yet, or a function returns nothing.

#### Advance and hold multiple values
- list → ordered, changeable collection that allows duplicates (example: ["apple", "banana", 10, True]). You can add, remove, change items. Very common for lists of things.
- tuple → ordered collection that cannot be changed after creation (example: (2025, "January", 15)). Used when data should stay fixed, like coordinates or days of the week.
- dict (dictionary) → unordered collection of key-value pairs (example: {"name": "Mehwish", "age": 25, "city": "Delhi"}). Keys are usually strings, values can be anything. Used for structured data like records or settings.
- set → unordered collection of unique values, no duplicates allowed (example: {1, 2, 3, "apple"}). Used when you want only unique items or to quickly check if something exists.

```python
#  int (whole number)
age = 25
print(age)          
print(type(age))    
```

```python
# float (decimal number)
height = 5.7
print(height)       
print(type(height)) 
```

```python
#  None (nothing / no value)
result = None
print(result)       
print(type(result)) 
```

```python
# 6. list (changeable collection)
fruits = ["apple", "banana", "mango"]
print(fruits)       
fruits.append("orange")
print(fruits)       
```

```python
# 7. tuple (unchangeable collection)
days = ("Monday", "Tuesday", "Wednesday")
print(days)        
# days[0] = "Sunday"   # ← this line would give error
```

```python
# 8. dict (key-value pairs)
person = {
    "name": "wish",
    "age": 22,
    "city": "Delhi"
}
print(person)              
print(person["name"])     
```

```python
# 9. set (unique values, no order, no duplicates)
numbers = {1, 2, 2, 3, 3, 4}
print(numbers)      
```
