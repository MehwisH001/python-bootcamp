Operators

Definition:-
- Operators are special symbols that perform operations on values (called operands).
- Example: + adds two numbers, == checks if two things are equal.

## arithmetic Operators- do math
- + add → 5 + 3 = 8
- - subtract → 10 - 4 = 6
- * multiply → 4 * 3 = 12
- / division → 10 / 2 = 5.0
- // floor division → 10 // 3 = 3
- % modulo (remainder) → 10 % 3 = 1
- ** exponent → 2 ** 3 = 8

## Assignment Operators— assign or update value
- = → x = 10
- += → x += 5 (same as x = x + 5)
- -=, *=, /=, //=, %=, **= (same idea)

## Comparison Operators- compare two values → give True/False
- == equal 
- != not equal 
- > greater than 
- < less than
- >= greater or equal
- <= less or equal

## Logical Operators- combine conditions (True/False)
- and → both must be True → True and False = False
- or → at least one True → True or False = True
- not → reverse → not True = False

## Identity Operators- check if same object in memory
- is → x is y (same object?)
- is not → opposite

## Membership Operators- check if value is inside collection
- in → "a" in "apple" → True
- not in → 5 not in [1,2,3] → True

## Bitwise Operators- work on binary (0s and 1s)used less by beginners
- & AND,  OR, ^ XOR, ~ NOT, << left shift, >> right shift

```python
#Comparison
print(10 > 5)     # True
print(7 == 7)     # True
print(3 != 8)     # True
print(4 <= 4)     # True
```

```python
# Arithmetic
print(10 + 5)    # 15
print(10 - 3)    # 7
print(4 * 3)     # 12
print(10 / 4)    # 2.5
print(10 // 4)   # 2
print(10 % 4)    # 2
print(2 ** 3)    # 8
```

```python
# Assignment
x = 20
x += 5       # x = x + 5 → 25
x -= 3       # 22
x *= 2       # 44
print(x)
```

```python
#  Logical
a = True
b = False
print(a and b)    # False
print(a or b)     # True
print(not a)      # False
```

```python
# Identity (most common use)
x = None
print(x is None)     # True
print(x is not None) # False
```
